## Task 1: Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
```

## Task 2: Create training set

```python
x = np.linspace(0, 1.571, 32)  # 32 values between 0 and pi/2
y = np.cos(x)  # cos(x) values
```

## Task 3: Normalize x to be between 0 and 1

```python
scaler = MinMaxScaler()
x_normalized = scaler.fit_transform(x.reshape(-1, 1))
```

## Task 4: Create Keras ANN

```python
model = Sequential([
    Dense(5, input_dim=1, activation='relu'),  # Layer 1 with 5 neurons
    Dense(3, activation='relu'),               # Layer 2 with 3 neurons
    Dense(1, activation='linear')             # Output layer with 1 neuron
])

# Compile the model
model.compile(optimizer='adam', loss='mse')
```

## Task 5: Train the ANN

```python
history = model.fit(x_normalized, y, epochs=3000, verbose=0)
```

## Task 6: Compare results

```python
predicted_y = model.predict(x_normalized)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='True Values', marker='o')
plt.plot(x, predicted_y, label='Predicted Values', marker='x')
plt.title('ANN Prediction vs. True Cosine Values')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()
plt.show()

# Evaluate training performance
final_loss = history.history['loss'][-1]
print(f"Final Training Loss: {final_loss}")
```
