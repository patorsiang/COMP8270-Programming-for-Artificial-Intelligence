## Task 1

Load the dataset `ClusterDs1.csv` from the Moodle page into a `DataFrame` and make a scatter plot using the `plt.scatter()` function.

```python
# load dataset ClusterDs1.csv into Dataframe
df = pd.read_csv('ClusterDs1.csv')
df
```

## Task 2

From the scatter plot of the previous exercise, you got a good idea of how many
cluster the data represents. Now create a K-Means model to find the clusters, and fit
it to the data points from the previous exercise.

```python
from sklearn.cluster import KMeans # required import
```

```python
km = KMeans(
    n_clusters = 3,   # number of clusters
    init = 'random',  # centroid initialisation
    n_init = 10,      # number of executions
    max_iter = 300,   # number of iterations
    random_state = 0  # random seed
)
```

```python
y_km = km.fit_predict(df)
```

```python
y_km
```

## Task 3

Using the model trained in the previous step, use the predict() method to get a list
of cluster labels. Assign the result to a variable named labels.

```python
df['labels'] = km.predict(df)
```

```python
df
```

## Task 4

Using the labels from the previous exercise, update the plot to show points in
different colour depending on what cluster they belong.

```python
# cluster 1
plt.scatter(
    df.loc[df['labels'] == 0, '0'],
    df.loc[df['labels'] == 0, '1'],
    s = 50,
    c = 'lightgreen',
    marker = 's',
    edgecolor = 'black',
    label = 'cluster 1'
)

# cluster 2
plt.scatter(
    df.loc[df['labels'] == 1, '0'],
    df.loc[df['labels'] == 1, '1'],
    s = 50,
    c = 'orange',
    marker = 'o',
    edgecolor = 'black',
    label = 'cluster 2'
)

# cluster 3
plt.scatter(
    df.loc[df['labels'] == 2, '0'],
    df.loc[df['labels'] == 2, '1'],
    s = 50,
    c = 'lightblue',
    marker = 'v',
    edgecolor = 'black',
    label = 'cluster 3'
)
```
