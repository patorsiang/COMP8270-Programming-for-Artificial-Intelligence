```python
import matplotlib.pyplot as plt   # Import the package
import math as math               # Only needed to generate some simple packages.  Not required by matplotlib

plt.rcParams['figure.figsize'] = [6, 6] # set plot size
plt.show ()
```

<p>_________________________________________________________________________________________________________</p>
<p style="font-size:25px">plot()</p>
<p style="font-size:20px">It is the simplest graph, but we can build on it.</p>
<p style="font-size:20px">Some uses include, but are not limited to:</p>
<ul style="font-size:20px">
    <li>Functions</li>
    <li>Time-Series</li>
    <li>Comparisons</li>
</ul>
    

```python
# Some data to play with, y = f(x) = x*x

x = [0.5, 1.5, 2.5, 3.5]
y = list (map (lambda x: x**2, x))
```

```python
plt.plot (y) # The simplest plot, just show me "y" please!

plt.show ()  # Render the graph
```

```python
plt.plot(x, y)    # Provide both x and y, x-axis is now accurate.

plt.show ()
```

<p>_________________________________________________________________________________________________________</p>
<p style="font-size:20px">The plt object is stateful.</p>
<p style="font-size:20px">We build graphs by adding elements and then rendering the plot.</p>
<p style="font-size:20px">We can add labels, titles - many things.</p>
<p style="font-size:20px">Once we have added all of our elements we use "show" to render and reset.</p>

```python
plt.plot(x, y, "orange")  # Add a colour   
plt.ylabel('y')           # Label for the y axis
plt.xlabel('x')           # Label for the x axis
plt.title ("f(x)")        # The plot title

plt.show ()               # Do it (state reset on exit)
```

```python
plt.ylabel('y')
plt.xlabel('x')
plt.title ("f(x)")

plt.plot(x, y, "ob")  # Blue points instead of a line

plt.show ()
```

```python
plt.ylabel('y')
plt.xlabel('x')
plt.title ("f(x)")

plt.plot(x, y, "gx")  # Green 'x'

plt.show ()
```

```python
plt.ylabel('y')
plt.xlabel('x')
plt.title ("f(x)")

plt.plot(x, y, "r-")  # Red line

plt.show ()
```

```python
plt.ylabel('y')
plt.xlabel('x')
plt.title ("f(x)")

plt.plot(x, y, "r-", x, y, "ro")  # Both points and a line

plt.show ()
```

```python
x = list (map (lambda x: x / 1.57, x))         # Convert to radians
y = list (map (lambda x: math.sin (x), x))     # sin (x)
z = list (map (lambda x: math.cos (x), x))     # cos (x)

# We can plot multiple data simultaneously

G = plt.plot (x, y, "bs", x, z, "r^")          # Confusing, the lines would help?

plt.show ()
```

```python
G = plt.plot (x, y, "bs", x, z, "r^", x, y, "g-", x, z, "r-") # This is unreadable and not very convenient

plt.show ()
```

```python
plt.xlabel ("radians")
plt.plot (x, y, "bs", x, y, "b-")      # y = sin (x)
plt.plot (x, z, "ro", x, z, "r-")      # z = cos (x)
plt.title ("Trig Functions")

plt.show ()
```

```python
import pandas as pd
VIX = pd.read_csv ("http://raptor.kent.ac.uk/~ds756/Data/VIX.csv") # VIX is a time-series data set
```

```python
plt.plot (VIX["Open"])  # Just the y, and there are many of them
plt.title ("This is very dense")

plt.show ()
```

```python
plt.plot ("Open", c="b", data=VIX)     # Now the plot knows about our Pandas dataframe, I can just use the names
plt.xlabel ("Time")
plt.ylabel ("Open")

plt.show ()
```

```python
plt.plot ("Date", "Open", "ro", data=VIX)
plt.plot ( "Close", "bs", data=VIX)
plt.xlabel ("Time")
plt.ylabel ("Volatility")

plt.show ()
```

```python
VIX_event = VIX[1000:1100]   # We could use data=VIX[1000:1100], but that is slower and more error prone

plt.plot ("Date", "Open", "ro", data=VIX_event)
plt.plot ("Date", "Open", "r-", data=VIX_event)
plt.plot ("Date", "Close", "bo", data=VIX_event)
plt.plot ("Date", "Close", "b-", data=VIX_event)

plt.title ("VIX from " + VIX_event["Date"].iloc[0] + " until " + VIX_event["Date"].iloc[99])
plt.xlabel ("Time")
plt.ylabel ("Volatility")

plt.show ()
```

<p style="font-size:25px">scatter()</p>
<p style="font-size:20px">It plots a set of points (x, y).</p>
<p style="font-size:20px">Some uses include, but are not limited to:</p>
<ul style="font-size:20px">
    <li>Relations</li>
    <li>First Glances</li>
    <li>More than 2 dimensions (x,y)</li>
</ul>


```python
heart = pd.read_csv ("http://raptor.kent.ac.uk/~ds756/Data/heart.csv")
```

```python
heart.columns
```

```python
plt.scatter ("RestingBP", "MaxHR", data = heart) # Examine the correlation between these two features
plt.xlabel ("RestingBP")
plt.ylabel ("MaxHR")

plt.show ()
```

```python
plt.scatter ("RestingBP", "MaxHR", c="Age", data = heart) # Use the age to set the colour
plt.xlabel ("RestingBP")
plt.ylabel ("MaxHR")

plt.show ()
```

```python
# Use cholesterol to set the size of the individual points
plt.scatter ("RestingBP", "MaxHR", c="Age", s="Cholesterol", data = heart) 

plt.xlabel ("RestingBP")
plt.ylabel ("MaxHR")

plt.show ()
```

```python
# Use cholesterol to set the size of the individual points
plt.scatter ("RestingBP", "MaxHR", c="HeartDisease", s="Cholesterol", data = heart) 

plt.xlabel ("RestingBP")
plt.ylabel ("MaxHR")

plt.show ()
```

```python
# Use cholesterol to set the size of the individual points
plt.scatter ("RestingBP", "MaxHR", c="HeartDisease", s=heart["Age"] * 1.5, data = heart) 

plt.xlabel ("RestingBP")
plt.ylabel ("MaxHR")

plt.show ()
```

```python
# We are not limited to what is in "heart", but lengths must match

Sizes = list( map (lambda x: x/5, heart.Cholesterol)) 

plt.scatter ("RestingBP", "MaxHR", c="Age", s=Sizes, data = heart)
plt.xlabel ("RestingBP")
plt.ylabel ("MaxHR")

plt.show ()
```

<p style="font-size:25px">bar()</p>
<p style="font-size:20px">Used for comparing non-numerical quantities (categorical).</p>
<p style="font-size:20px">Some uses include, but are not limited to:</p>
<ul style="font-size:20px">
    <li>Comparing classes and quantities</li>
    <li>Can be used for a histogram</li>
</ul>



```python
# Create a small dataframe to demonstrate classes and values

SalesDict = {
    "GEO0" : { "Geo" : "Kent", "Sales" : 100}, 
    "GEO1" : { "Geo" : "Surrey", "Sales" : 110},
    "GEO2" : { "Geo" : "Essex", "Sales" : 110},
    "GEO3" : { "Geo" : "London", "Sales" : 120}
}

Sales = pd.DataFrame(data=SalesDict).T

plt.bar (Sales.Geo, Sales.Sales)

plt.show ()
```

```python
# max (Freq) = mode

Freq = heart[["ChestPainType"]].groupby (["ChestPainType"]).size ()
Classes = Freq.index
Classes
```

```python
plt.bar (Classes, Freq)
plt.xlabel ("Type of Chest Pain")
plt.ylabel ("Frequency")
plt.title ("I'm a histogram!")

plt.show ()
```

<p>_________________________________________________________________________________________________________</p>
<p style="font-size:25px">hist()</p>
<p style="font-size:20px">Used for plotting distributions.</p>
<p style="font-size:20px">Some uses include, but are not limited to:</p>
<ul style="font-size:20px">
    <li>Estimating emprical densities</li>
    <li>Discrete empirical distributions</li>
    <li>More than 2 dimensions (x,y)</li>
</ul>




```python
plt.hist (heart["ChestPainType"])   # I do the work of computing frequencies for you

plt.show ()
```

```python
plt.hist (heart["Age"])
plt.xlabel ("Age")
plt.ylabel ("Frequency")
plt.grid (True)

plt.show ()
```

```python
plt.hist (heart["Age"], 25)   # Increase the number of bins
plt.xlabel ("Age")
plt.ylabel ("Frequency")
plt.grid (True)

plt.show ()
```

```python
plt.hist (heart["Age"])
plt.xlabel ("Age")
plt.ylabel ("Frequency")
plt.grid (True)

# We can annotate our graph with text
plt.text (30, 177, "The mode is close!") 

plt.show ()
```

```python
plt.hist (heart["Age"])
plt.xlabel ("Age")
plt.ylabel ("Frequency")
plt.grid (True)

# We can annotate our graph with objects
plt.annotate ("mode", xy=(51, 200), xytext=(30, 160), arrowprops=dict(facecolor="red")) 

plt.show ()
```

```python
plt.hist (heart["Age"], 20, density = 1, facecolor="c") # Density function instead of a histogram
plt.xlabel ("Age")
plt.ylabel ("Frequency")
plt.grid (True)
```
